```python
import pandas as pd
import tensorflow as tf
train =pd.read_csv("C:/Users/Lab009/Desktop/dataset/train.csv")
test =pd.read_csv("C:/Users/Lab009/Desktop/dataset/test.csv")

print(train.shape)
print(test.shape)
x=train.drop('label',axis=1).values
y=train[['label']].values

from sklearn.model_selection import  train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
xtrain=tf.keras.utils.normalize(xtrain)
xtest=tf.keras.utils.normalize(xtest)
model=tf.keras.models.Sequential()
#input layers
model.add(tf.keras.layers.Flatten())
#model.add(tf.keras.layers.Dense(64))
#hidden layers
model.add(tf.keras.layers.Dense(64,activation='relu'))
#hidden layers
model.add(tf.keras.layers.Dense(128,activation='relu'))
#output layers
model.add(tf.keras.layers.Dense(10,activation='softmax'))

#activation function
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(xtrain,ytrain,epochs=9)

pred=model.predict(xtest)
```

```python
pred.shape
pv=[]
import numpy as np
for i in range(len(pred)):
    pv.append(np.argmax(pred[i]))
#print(pv)

pv=np.array(pv)
pv
```

```python
pv.shape
test['predict']=pd.DataFrame({'predict':pv})
```
