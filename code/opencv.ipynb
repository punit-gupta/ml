```python
def label(img):
    return img.split('.')[0]

def label1(img):
    l=img.split('.')[0]
    if(l=='cat'):
        return 0
    elif(l=='dog'):
        return 1
```

```python
#label(a)
label1(a)

```

```python
training=[]
def training_data():
    for a in tqdm(os.listdir(train)):
        l=label1(a)
        path=os.path.join(train,a)
        f=cv2.imread(path,0)# 0 refer to gray scle imge rather than color
        cv2.resize(f,(50,50))
        training.append([[f],[l]])
    shuffle(training)
    return training
        
```
